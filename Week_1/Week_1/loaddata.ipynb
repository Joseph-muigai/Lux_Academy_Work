```python
# Find all records where the weather was exactly clear.
data[data['Weather'] == 'Clear']
```

```python
# Find the number of times the wind speed was exactly 4 km/hr.
data[data['Wind Speed_km/h'] == 4 ]
```

```python
# Check if there are any NULL values present in the dataset.
data.isnull()
```

```python
# Rename the column "Weather" to "Weather_Condition."
data.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
```

```python
# What is the mean visibility of the dataset?
data['Visibility_km'].mean()
```

```python
# Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]
```

```python
data = data.apply(pd.to_numeric, errors='coerce')
data.groupby('Weather_Condition').mean()
```

```python
#Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
data[(data['Weather_Condition'] == 'Clear') & ((data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40))]



```

```python
# Find the number of weather conditions that include snow.
data['Weather_Condition'].str.contains('Snow').sum()
```

```python
import csv
import mysql.connector

# MySQL connection parameters
cnx = mysql.connector.connect(user='root', password='1234', host='localhost', database='weatherdata')
cursor = cnx.cursor()
cursor.execute("CREATE TABLE weatherdata1 (`Date/Time` VARCHAR(255), Temp_C FLOAT, `Dew Point Temp_C` FLOAT, `Rel Hum_%` FLOAT, `Wind Speed_km/h` FLOAT, `Visibility_km` FLOAT, `Press_kPa` FLOAT, `Weather_Condition` VARCHAR(255))")

# CSV file path
csv_file_path = '1. Weather Data.csv'

with open(csv_file_path, mode='r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Skip the header row

    for row in reader:
        cursor.execute("INSERT INTO weatherdata1 (`Date/Time`, `Temp_C`, `Dew Point Temp_C`, `Rel Hum_%`, `Wind Speed_km/h`, `Visibility_km`, `Press_kPa`, `Weather_Condition`) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)", row)

cnx.commit()
cursor.close()
cnx.close()
```
